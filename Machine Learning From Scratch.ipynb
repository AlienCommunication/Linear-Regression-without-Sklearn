```python
import pandas as pd
import numpy as np
from linear_regression_model import linear_regression_model
```

```python
data = pd.read_csv("train-data.csv")
```

```python
data.head()
```

```python
#Spliting the dataset into X and Y form
X = data.drop(['Price', 'New_Price', "Unnamed: 0"], axis = 1)
```

```python
X.head()
```

```python
y = data['Price']
```

```python
y.head()
```

```python
# Checking for dtypes

pd.options.display.max_rows = 65

dtype = X.dtypes.reset_index()
dtype.columns = ["Variable Name", "Variable Type"]
dtype

```

```python
# Cleaning the variables-
#Look at our Name variable- It has Brand name and Model name into just one column hence I will split it into two new columns "Brand" and "Model"
names = list(X.Name)
brand = []
model = []
for i in range(len(names)):
    try:
        brand.append(names[i].split(" ")[0].strip())
        try:
            model.append(" ".join(names[i].split(" ")[1:]).strip())
        except:
            pass
    except:
        print("ERR ! - ", names[i], "@" , i)
X["Brand"] =  brand
X["Model"] = model
```

```python
#Now There is texts in the Mileage, Engine, Power- I will remove the text and keep the relevel information only, This operation will be applied to every rows except 
#Those will null values 

Engine = list(X.Engine)
for i in range(len(Engine)):
    try :
        Engine[i] = float(Engine[i].split(" ")[0].strip())
    except:
        Engine[i] = np.nan
X['Engine'] = Engine
```

```python
Power = list(X.Power)
for i in range(len(Power)):
    try :
        Power[i] = float(Power[i].split(" ")[0].strip())
    except:
        Power[i] = np.nan
X['Power'] = Power
```

```python
mileage = list(X.Mileage)
for i in range(len(mileage)):
    try :
        mileage[i] = float(mileage[i].split(" ")[0].strip())
    except:
        mileage[i] = np.nan
X['Mileage'] = mileage
```

```python
X.head()
```

```python
#Still name column is there in the data hence will remove it- 

X=X.drop(['Name'], axis = 1)
```

```python
X.head()
```

```python
# Checking for Missing values-
def get_percentage_missing(series):
    num = series.isnull().sum()
    den = len(series)
    return round(num/den, 2)
# Only include columns that contain any NaN values
df_with_any_null_values = X[X.columns[X.isnull().any()].tolist()]

get_percentage_missing(df_with_any_null_values)
```

```python
#Checking Maximum values in every columns including the Nan

maxValuesObj = X.max(skipna=False)
 
print('Maximum value in each column including NaN: ')
print(maxValuesObj)
```

```python
#Checking Minimum  values in every columns including the Nan

minValuesObj = X.min(skipna=False)
 
print('Minimum value in each column including NaN: ')
print(minValuesObj)
```

```python
# creating bool series True for NaN values   - I will do this for every columns to find the How does the NaN values look like
# Hence I will be changing the Column here in the same code to get objective done and will write the result below
bool_series = pd.isnull(X["Engine"])  
    
# filtering data  
# displaying data only with Gender = NaN  
X[bool_series]
```

```python
#Let's calculate the mean for these columns to get the nan values replaced by mean.

X[["Engine", "Mileage", "Power", "Seats"]].mean()
```

```python
X['Mileage']=X['Mileage'].fillna(X['Mileage'].mean())

X['Engine']=X['Engine'].fillna(X['Engine'].mean())


X['Power']=X['Power'].fillna(X['Power'].mean())

X['Seats']=X['Seats'].fillna(X['Seats'].mean())
```

```python
# Scaling the Variables 


#def normalize(X.Engine):
    #return [round((i - min(X.Engine)) / (max(X.Engine) - min(X.Engine)), 2) for i in X.Engine]

#print(normalize(X))
```

```python
#Encoding Categorical into Numerical form

Encoded_Data= pd.get_dummies(data=X, columns=['Location', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Brand'], drop_first=True)
```

```python
Encoded_Data.head()
```

```python
Encoded_Data=Encoded_Data.drop(['Model'], axis=1)
```

```python
Encoded_Data.dtypes
```

```python
#Normalizing the matrix between -1 and +1

X = (Encoded_Data - Encoded_Data.mean())/(Encoded_Data.max() - Encoded_Data.min())
```

```python
X.head()
```
